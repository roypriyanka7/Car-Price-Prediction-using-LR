```python
# encoding ordinal categorical columns

data['cylindernumber'] = data['cylindernumber'].map({'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12})
```

```python
# creating features and label variable

X= data[['wheelbase','carlength', 'carwidth', 'boreratio', 'stroke', 'curbweight',
       'cylindernumber', 'enginesize','horsepower', 'peakrpm', 'compressionratio' ]].copy()
y = data['price']

# print(X)
# print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 32)
```

```python
# scaling data

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
auc = lr.score(X_test, y_test)
print('Accuracy of the model is : ', (auc * 100).round(2), '%')
```
